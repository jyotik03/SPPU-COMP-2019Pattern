```python
x=2
lr=0.0067
max_item=10067
iter=0
precision=0.000001
previous_step_size=1
gf=lambda x:(x+3)**2
```

```python
gd=[]
while precision < previous_step_size and iter<max_item:
    prev=x
    x=x-lr*gf(prev)
    previous_step_size=abs(x-prev)
    iter+=1
    print("Iterations : ",iter," ","Values : ",x)
    gd.append(x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(gd)
```
